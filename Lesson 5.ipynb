```python
M = int (input('M='))
N = int (input('N='))

def print_matrix_elements(M, N):
    matrix = [[(i * N) + j + 1 for j in range(N)] for i in range(M)]

    for j in range(N):
        if j % 2 == 0:
            for i in range(M):
                print(matrix[i][j], end=' ')
        else:
            for i in range(M-1, -1, -1):
                print(matrix[i][j], end=' ')
        print()

print_matrix_elements(M, N)
```

```python
M = int (input('M='))
N = int (input('N='))

def sum_of_rows(M, N):
    matrix = [[(i * N) + j + 1 for j in range(N)] for i in range(M)]
    row_sums = [sum(row) for row in matrix]
    return row_sums

row_sums = sum_of_rows(M, N)
print("Суммы элементов каждой строки матрицы:")
for i, row_sum in enumerate(row_sums):
    print(f"Сумма элементов строки {i+1}: {row_sum}")
```

```python
M = int (input('M='))
N = int (input('N='))
def find_max_sum_row(M, N):
    matrix = [[(i * N) + j + 1 for j in range(N)] for i in range(M)]

    max_sum = float('-inf')
    max_sum_row = None

    for row in matrix:
        current_sum = sum(row)
        if current_sum > max_sum:
            max_sum = current_sum
            max_sum_row = row
    return max_sum_row, max_sum

max_sum_row, max_sum = find_max_sum_row(M, N)
print("Строка с наибольшей суммой элементов:")
print("Строка:", max_sum_row)
print("Сумма:", max_sum)
```

```python
M = int (input('M='))
N = int (input('N='))
def find_min_max_columns(M, N):
    matrix = [[(i * N) + j + 1 for j in range(N)] for i in range(M)]

    min_elements = []
    max_elements = []

    for j in range(N):
        column = [matrix[i][j] for i in range(M)]
        min_elements.append(min(column))
        max_elements.append(max(column))
    return min_elements, max_elements

min_elements, max_elements = find_min_max_columns(M, N)
print("Минимальные элементы каждого столбца:")
print(min_elements)
print("Максимальные элементы каждого столбца:")
print(max_elements)
```
