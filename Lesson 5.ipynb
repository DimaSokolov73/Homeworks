```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def find_mersenne_primes(n):
    mersenne_primes = []
    p = 2
    while True:
        k = (2 ** p) - 1
        if k >= n:
            break
        if is_prime(p):
            mersenne_primes.append (k)
        p += 1
    return mersenne_primes

n = int(input("Введите натуральное число n: "))

if n > 0:
    mersenne_primes = find_mersenne_primes(n)
    print(f"Числа Мерсена, меньшие {n}: {mersenne_primes}")
else:
    print("Число n должно быть натуральным (больше 0).")
```

```python
def calculate_averages(A):
    B = [(A[i] + A[i + 1]) / 2
    for i in range(0, len(A), 2)]
    return B

A = [1, 3, 5, -2, 0, 4, 0, 3]
if len(A) % 2 == 0:
    B = calculate_averages(A)
    print(f"Массив B: {B}")
else:
    print("Длина массива A должна быть четной.")
```

```python
import math
def LongInt(x):
    return math.factorial(x)

def calculate_sum(n):
    total_sum = 0
    for i in range(n):
        z = (2 * i + 1)
        if i % 2 == 0:
            total_sum += LongInt(z)
        else:
            total_sum -= LongInt(z)
    return total_sum

n = int(input("Введите четное число n (n ≤ 20): "))

if n <= 20 and n % 2 == 0:
    result = calculate_sum(n)
    print(f"Сумма ряда: {result}")
else:
    print("Число n должно быть четным и не больше 20.")
```

```python
def is_power_of_five(K):
    if K <= 0:
        return False
    while K % 5 == 0:
        K //= 5
    return K == 1

def count_powers_of_five(numbers):
    count = 0
    for number in numbers:
        if is_power_of_five(number):
            count += 1
    return count

numbers = [5, 25, 125, 625, 3125, 10, 20, 30, 40, 50]

count = count_powers_of_five(numbers)
print(f"Количество степеней числа 5 в наборе: {count}")
```

```python
def Fact(N):
    if N < 0:
        return
    result = 1.0
    for i in range(1, N + 1):
        result *= i
    return result

numbers = [1, 2, 3, 4, 5]

for number in numbers:
    factorial = Fact(number)
    print(f"Факториал {number}! = {factorial}")

```
