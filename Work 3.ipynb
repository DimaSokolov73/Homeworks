```python
import math
N = int(input('N='))
y = 1/math.cos(1)
k = math.cos(1)
for i in range (2,N+1):
  k = k+math.cos(i)
  y = y+(1/(k))
print (y)
```

```python
n = float (input ('n='))
epsilon=0.001
a= float (input ('a='))
b= float (input ('b='))
sum=0
while sum>=epsilon:
  K= 1/a**(n)+1/b**(n)
  sum=sum+K
print (K)
```

```python
N = int(input("Введите количество чисел N (> 1): "))
numbers = list(map(float, input("Введите N вещественных чисел через пробел: ").split()))
for i in range(1, N):
    if numbers[i] > numbers[i - 1]:
        print(f"Нарушение закономерности на числе с номером {i + 1}")
        break
else:
    print("0")




```
